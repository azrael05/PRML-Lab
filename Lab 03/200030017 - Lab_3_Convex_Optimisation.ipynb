# Find the value of $x$ at which $f(x)$ is minimum : 
1. Find $x$ analytically
2. Write the update equation of gradient descent
3. Find $x$ using gradient descent method 



**Example 1** : $f(x)$ = $x^{2}$ + $x$ + $2$

**Analytical** : $$\frac{d}{dx} f(x)=2x+1=0$$
$$\frac{d^{2}}{dx^{2}} f(x)=2~(Minima)$$
$$x=-\frac{1}{2}~(analytical~solution)$$

**Gradient Descent Update equation** : $$x_{init}=4$$
$$x_{updt}=x_{old}-\lambda (\frac{d}{dx} f(x)|x=x_{old})$$
$$x_{updt}=x_{old}-\lambda (2x_{old}+1)$$

**Gradient Descent Method** :

Follow the below steps and write your code in the block below

1. Generate $x$, 1000 data points from -10 to 10
2. Generate and Plot the function $f(x)$ = $x^{2}$ + $x$ + $2$
3. Initialize the starting point ($x_{init}$) and learning rate ($\lambda$)
4. Use Gradient descent algorithm to compute value of $x$ at which the function $f(x)$ is minimum
5. Also vary the learning rate and initialisation point and plot your observations

```python
## Write your code here

## Define gradient function
def gradient(x):
    return 2*x+1

## Generate datapoints
x=np.linspace(start=-10,stop=10,num=1000)
fig=plt.figure(figsize=(5,5),edgecolor="red")

#plot the function
y=x*x+x+2
plt.plot(x,y)

#give initial value and learning rate
x_init=-10
learning_rate=0.8
x_values=[]
x_values.append(x_init)
for i in range(1000):
    x_upt=x_init-learning_rate*gradient(x_init)
    x_values.append(x_upt)
    if(round(x_upt,10)==round(x_init,10)):
        break
    x_init=x_upt
print("Position for minimum cost with initial point(x)-(",x_values[0],") and learning rate-",learning_rate," is ",x_init)
x_values=np.array(x_values)

# plot gradient descent
y_values=x_values*x_values+x_values+2
plt.plot(x_values,y_values,'red')
plt.xlim(-11,11)
plt.ylim(-5,150)
plt.show()



#plot the function
y=x*x+x+2
plt.plot(x,y)

#give initial value and learning rate
x_init=-10
learning_rate=0.05
x_values=[]
x_values.append(x_init)
for i in range(1000):
    x_upt=x_init-learning_rate*gradient(x_init)
    x_values.append(x_upt)
    if(round(x_upt,10)==round(x_init,10)):
        break
    x_init=x_upt
print("Position for minimum cost with initial point(x)-(",x_values[0],") and learning rate-",learning_rate," is ",x_init)
x_values=np.array(x_values)

# plot gradient descent
y_values=x_values*x_values+x_values+2
plt.plot(x_values,y_values,'red')
plt.xlim(-11,11)
plt.ylim(-5,150)
plt.show()



#plot the function
y=x*x+x+2
plt.plot(x,y)

#give initial value and learning rate
x_init=4
learning_rate=0.05
x_values=[]
x_values.append(x_init)
for i in range(1000):
    x_upt=x_init-learning_rate*gradient(x_init)
    x_values.append(x_upt)
    if(round(x_upt,10)==round(x_init,10)):
        break
    x_init=x_upt
print("Position for minimum cost with initial point(x)-(",x_values[0],") and learning rate-",learning_rate," is ",x_init)
x_values=np.array(x_values)

# plot gradient descent
y_values=x_values*x_values+x_values+2
plt.plot(x_values,y_values,'red')
plt.xlim(-11,11)
plt.ylim(-5,150)
plt.show()

```

#Find the value of $x$ and $y$ at which $f(x,y)$ is minimum :



**Example 1** : $f(x,y)$ = $x^{2}+y^{2}+2x+2y$

**Gradient Descent Method** :

Follow the below steps and write your code in the block below

1. Generate $x$ and $y$, 1000 data points from -10 to 10
2. Generate and Plot the function $f(x,y)$ = $x^{2}+y^{2}+2x+2y$
3. Initialize the starting point ($x_{init}$,$y_{init}$) and learning rate ($\lambda$)
4. Use Gradient descent algorithm to compute value of $x$ and $y$ at which the function $f(x,y)$ is minimum
5. Also vary the learning rate and initialisation point and plot your observations

```python
## Write your code here (Ignore the warning)

## Define gradient functions
def gradient_x(x,y):
    return 2*x+2
def gradient_y(x,y):
    return 2*y+2

# Generate data points
x=np.linspace(-10,10,1000)
y=np.linspace(-10,10,1000)
[X,Y]=np.meshgrid(x,y)

#plot the function
z=pow(X,2)+pow(Y,2)+2*X+2*Y                                   
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,z,color='blue')
plt.figure()

#plot the contour
ax1=plt.axes(projection="3d")
ax1.contour(X,Y,z,50)

#Give initial values 
x_init=2
y_init=2
learning_rate=0.8
x=x_init
y=y_init

#plot the gradient descent
plt.figure()
plt.contour(X,Y,z,100)
plt.title("x="+str(x)+", y="+str(y)+", lr="+str(learning_rate))

#Perform gradient descent 
while(1):
    x_upt=x_init-learning_rate*gradient_x(x_init,y_init)
    y_upt=y_init-learning_rate*gradient_y(x_init,y_init)
    plt.plot([x_init,x_upt],[y_init,y_upt],color='k')   # plotting  learning rate for different initial points in graph
    if((round(x_upt,10)==round(x_init,10)) and (round(y_upt,10)==round(y_init,10))):
        break
    x_init=x_upt
    y_init=y_upt


#Print the values for minima
print("Position for minimum cost with initial point(x,y)-(",x,",",y,") and learning rate-",learning_rate," is ",x_init,y_init)




#Give initial values 
x_init=2
y_init=2
learning_rate=0.05
x=x_init
y=y_init

#plot the gradient descent
plt.figure()
plt.contour(X,Y,z,100)
plt.title("x="+str(x)+", y="+str(y)+", lr="+str(learning_rate))

#Perform gradient descent 
while(1):
    x_upt=x_init-learning_rate*gradient_x(x_init,y_init)
    y_upt=y_init-learning_rate*gradient_y(x_init,y_init)
    plt.plot([x_init,x_upt],[y_init,y_upt],color='k')   # plotting  learning rate for different initial points in graph
    if((round(x_upt,10)==round(x_init,10)) and (round(y_upt,10)==round(y_init,10))):
        break
    x_init=x_upt
    y_init=y_upt


#Print the values for minima
print("Position for minimum cost with initial point(x,y)-(",x,",",y,") and learning rate-",learning_rate," is ",x_init,y_init)




#Give initial values 
x_init=4
y_init=4
learning_rate=0.05
x=x_init
y=y_init

#plot the gradient descent
plt.figure()
plt.contour(X,Y,z,100)
plt.title("x="+str(x)+", y="+str(y)+", lr="+str(learning_rate))

#Perform gradient descent 
while(1):
    x_upt=x_init-learning_rate*gradient_x(x_init,y_init)
    y_upt=y_init-learning_rate*gradient_y(x_init,y_init)
    plt.plot([x_init,x_upt],[y_init,y_upt],color='k')   # plotting  learning rate for different initial points in graph
    if((round(x_upt,10)==round(x_init,10)) and (round(y_upt,10)==round(y_init,10))):
        break
    x_init=x_upt
    y_init=y_upt


#Print the values for minima
print("Position for minimum cost with initial point(x,y)-(",x,",",y,") and learning rate-",learning_rate," is ",x_init,y_init)



```
